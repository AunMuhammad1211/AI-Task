```python
class TodoList:
    def __init__(self):
        self.task = []

    def add_task(self,task):    # for add task
        self.task.append({"Task": task, "Completed": False})

    def remove_task(self,index):
        if 0 <= index < len(self.task):
            self.task.pop(index)
        else:
            print("Index Error: Index not find.")

    def update_task(self,index,new_value):
        if 0 <= index < len(self.task):
            self.task[index] = {"Task": new_value, "Completed": False}
        else:
            print("Index Error: Index out of range.")

    def mark_as_completed(self,index):
        if 0 <= index < len(self.task):
            self.task[index]["Completed"] = True
        else:
            print("Index Error. Index not found.")

    def display_tasks(self):
        if not self.task:
            print("List is empty.Add task to it.")
        else:
            for i,task in enumerate(self.task,start=1):
                status = "Completed" if task["Completed"] else "Pending"
                print(f"{i}.{task} --> {status}")
def main():
    todolist = TodoList()
    while True:
        print("1. Add Task to the list")
        print("2. Remove Task from list")
        print("3. Update Task from the list")
        print("4. Mark as completed")
        print("5. Display the all Tasks")
        print("6. Exit")
        choice = int(input("Enter the input to perform your specific choice: "))

        if choice == 1:
            task = input("Enter the task: ")
            todolist.add_task(task)

        elif choice == 2:
            index = int(input("Enter the index of the value you want to remove: "))
            todolist.remove_task(index)

        elif choice == 3:
            index = int(input("Enter the index value where you want to insert new task: "))
            new_value = input("Enter the new task to add to the list: ")
            todolist.update_task(index,new_value)

        elif choice == 4:
            todolist.display_tasks()
            index = int(input("Enter the index of the task you want to mark as completed: "))
            todolist.mark_as_completed(index)

        elif choice == 5:
            todolist.display_tasks()

        elif choice == 6:
            print("Exit! Thanks for playing this game.")
            break
        else:
            print("Invalid choice!")


if __name__ == '__main__':
    main()

```
