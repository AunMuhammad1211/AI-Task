# Data Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder
```

# Encoding

```python
ordinal_encoder = OrdinalEncoder(categories=[
    ['Fair', 'Good', 'Very Good', 'Ideal', 'Premium'],  # cut
    ['D', 'E', 'F', 'G', 'H', 'I', 'J'],               # color
    ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']  # clarity
])
df1[['cut', 'color', 'clarity']] = ordinal_encoder.fit_transform(df1[['cut', 'color', 'clarity']])
```

```python
df1.drop(columns=["cut","color"])
```

# Now we convert the remaining values of column to same scale for easy calulation and accurate solution

```python
def MinMaxNormalize(columns):
	for column in columns:
		datatypes = df1[column].dtypes
		if datatypes != "object":
			max_value = df1[column].max()
			min_value = df1[column].min()
			if max_value != min_value:
				df1[column] = (df1[column] - min_value) / (max_value - min_value)
        
columns = df1.columns
print(columns)
MinMaxNormalize(columns)
```

```python
df1
```

# Column without target variable

```python
X = df1.drop(columns=["price","cut","color","clarity"])
print(X)
```

# Target Variable

```python
Y = df1["price"]
print(Y)
```

# Training and testing of data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

# Train the model by SVM Classfier

```python
from sklearn.svm import SVR
```

```python
model = SVR(kernel='linear')
model.fit(X_train, Y_train)
```

# Evaluate the model

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_predict = model.predict(X_test)
mse = mean_squared_error(Y_test,y_predict )
print(f"Mean Squared Error: {mse}")


mae = mean_absolute_error(Y_test, y_predict)
print(f"Mean absolute error: {mae}")


r2_s = r2_score(Y_test, y_predict)
print(f"r2 score: {r2_s*100}")
```
